```python
import re
import pandas as pd

def extract_passing_lines(pass_input):
    # Regex pattern to capture the player's first name, last name, team, and "Over" value
    pattern = re.compile(
        r"(?P<first_name>\w+)\n"           # First name
        r"(?P<last_name>[^\n]+)\n"         # Last name
        r"(?P<team>[^\n]+)\n"              # Team name
        r"[^\n]+\n[^\n]+\n@\n[^\n]+\n"     # Skip match details
        r"Total Passing Yards Thrown\n\n"
        r"O (?P<over_value>[\d.]+)"        # "Over" value
    )
    
    # List to store each player's information as dictionaries
    player_data = []

    # Iterate over all matches and extract necessary data
    for match in pattern.finditer(pass_input):
        first_name = match.group("first_name")
        last_name = match.group("last_name")
        team = match.group("team")
        over_value = float(match.group("over_value"))  # Convert over value to float
        
        # Debugging: Print each extracted match
        # print(f"Match found: {first_name} {last_name}, Team: {team}, Over Value: {over_value}")

        # Append data as a dictionary
        player_data.append({
            "Player": f"{first_name} {last_name}",
            "Team": team,
            "Over Value": over_value
        })

    # Check if player data is populated
    if not player_data:
        print("No matches found in the input text.")

    # Create a DataFrame from the list of dictionaries
    pass_df = pd.DataFrame(player_data)

    pass_df.columns = ['player','team','line']
    pass_df['market'] = 'passing_yards'
    pass_df = pass_df[['player','market','line']]


    return pass_df

# rushing
def extract_rushing_lines(rush_input):
    # Regex pattern to capture the player's first name, last name, team, and "Over" value
    pattern = re.compile(
        r"(?P<first_name>\w+)\n"           # First name
        r"(?P<last_name>[^\n]+)\n"         # Last name
        r"(?P<team>[^\n]+)\n"              # Team name
        r"[^\n]+\n[^\n]+\n@\n[^\n]+\n"     # Skip match details
        r"Total Rushing Yards Recorded\n\n"
        r"O (?P<over_value>[\d.]+)"        # "Over" value
    )
    
    # List to store each player's information as dictionaries
    player_data = []

    # Iterate over all matches and extract necessary data
    for match in pattern.finditer(rush_input):
        first_name = match.group("first_name")
        last_name = match.group("last_name")
        team = match.group("team")
        over_value = float(match.group("over_value"))  # Convert over value to float
        
        # Debugging: Print each extracted match
        # print(f"Match found: {first_name} {last_name}, Team: {team}, Over Value: {over_value}")

        # Append data as a dictionary
        player_data.append({
            "Player": f"{first_name} {last_name}",
            "Team": team,
            "Over Value": over_value
        })

    # Check if player data is populated
    if not player_data:
        print("No matches found in the input text.")

    # Create a DataFrame from the list of dictionaries
    rush_df = pd.DataFrame(player_data)

    rush_df.columns = ['player','team','line']
    rush_df['market'] = 'rushing_yards'
    rush_df = rush_df[['player','market','line']]

    return rush_df

# receiving
def extract_receiving_lines(rec_input):
    # Regex pattern to capture the player's first name, last name, team, and "Over" value
    pattern = re.compile(
        r"(?P<first_name>\w+)\n"           # First name
        r"(?P<last_name>[^\n]+)\n"         # Last name
        r"(?P<team>[^\n]+)\n"              # Team name
        r"[^\n]+\n[^\n]+\n@\n[^\n]+\n"     # Skip match details
        r"Total Yards Received\n\n"
        r"O (?P<over_value>[\d.]+)"        # "Over" value
    )
    
    # List to store each player's information as dictionaries
    player_data = []

    # Iterate over all matches and extract necessary data
    for match in pattern.finditer(rec_input):
        first_name = match.group("first_name")
        last_name = match.group("last_name")
        team = match.group("team")
        over_value = float(match.group("over_value"))  # Convert over value to float
        
        # Debugging: Print each extracted match
        # print(f"Match found: {first_name} {last_name}, Team: {team}, Over Value: {over_value}")

        # Append data as a dictionary
        player_data.append({
            "Player": f"{first_name} {last_name}",
            "Team": team,
            "Over Value": over_value
        })

    # Check if player data is populated
    if not player_data:
        print("No matches found in the input text.")

    # Create a DataFrame from the list of dictionaries
    rec_df = pd.DataFrame(player_data)

    # clean rec input
    rec_df.columns = ['player','team','line']
    rec_df['market'] = 'receiving_yards'
    rec_df = rec_df[['player','market','line']]

    return rec_df
```
